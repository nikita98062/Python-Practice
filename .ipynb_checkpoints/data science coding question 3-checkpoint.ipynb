```python
# PCA for Noise Reduction

from sklearn.datasets import load_digits
```

```python
df=load_digits()
```

```python
df
```

```python
x=df.data
```

```python
x
```

```python
noise_factor=0.5
```

```python
x_noise=x+noise_factor*np.random.normal(loc=0.0,scale=1.0,size=x.shape)
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=20,random_state=42)
```

```python
x_pca=pca.fit_transform(x_noise)
```

```python
x_inverse=pca.inverse_transform(x_pca)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].imshow(x[0].reshape(8, 8), cmap='gray')
axes[0].set_title("Original Image")

axes[1].imshow(x_noise[0].reshape(8, 8), cmap='gray')
axes[1].set_title("Noisy Image")

axes[2].imshow(x_inverse[0].reshape(8, 8), cmap='gray')
axes[2].set_title("Denoised Image using PCA")

plt.show()
```

```python
# ANOVA (Analysis of Variance)
import pandas as pd
import seaborn as sns
from scipy.stats import f_oneway
import matplotlib.pyplot as plt
```

```python
data = {
    'Region': ['North']*5 + ['South']*5 + ['East']*5,
    'Sales': [220, 250, 230, 240, 210, 180, 190, 175, 200, 185, 260, 270, 250, 245, 255]
}
df = pd.DataFrame(data)
df
```

```python
#ANOVA Test
north = df[df['Region'] == 'North']['Sales']
south = df[df['Region'] == 'South']['Sales']
east = df[df['Region'] == 'East']['Sales']
f_stat,p_val=f_oneway(north,south,east)
f_stat,p_val
```

```python
# Visualization
sns.boxplot(x='Region', y='Sales', data=df)
plt.title('Sales Distribution by Region')
plt.show()
```

```python
# Topic: Hypothesis Testing (t-test)
from scipy.stats import ttest_ind
import seaborn as sns
```

```python
df = pd.DataFrame({
    'Gender': ['Male']*5 + ['Female']*5,
    'Spending': [250, 270, 230, 260, 245, 220, 210, 225, 235, 215]
})
df
```

```python
male_spend = df[df['Gender'] == 'Male']['Spending']
female_spend = df[df['Gender'] == 'Female']['Spending']
```

```python
#Topic: Mann-Whitney U Test (Non-parametric)
from scipy.stats import mannwhitneyu

df = pd.DataFrame({
    'Product': ['A']*5 + ['B']*5,
    'Satisfaction': [4, 5, 4, 3, 5, 2, 3, 2, 1, 3]
})

prod_a = df[df['Product'] == 'A']['Satisfaction']
prod_b = df[df['Product'] == 'B']['Satisfaction']

stat, p_val = mannwhitneyu(prod_a, prod_b)
print("U-Statistic:", stat)
print("P-Value:", p_val)

sns.boxplot(x='Product', y='Satisfaction', data=df)
plt.title("Satisfaction Score Comparison")
plt.show()
```
