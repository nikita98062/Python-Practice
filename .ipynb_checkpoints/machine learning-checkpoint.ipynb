```python
#  Question 4: Use elbow method to choose k (customer spending)
import pandas as pd
df = pd.DataFrame({
    'AnnualIncome': [15,16,17,18,19,20,40,42,43,44,45,70,72,73,74,75,80,82,85,90],
    'SpendingScore': [39,81,6,77,40,76,6,94,3,72,10,40,42,39,50,60,70,85,90,88]
})
df
```

```python
# use elbow method to find optimal k
# fit Kmeans and visualize clusters

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Elbow method
inertia=[]
elbow=range(1,7)
for k in elbow:
    model=KMeans(n_clusters=k,random_state=1)
    model.fit(df)
    inertia.append(model.inertia_)
```

```python
model.inertia_
```

```python
import seaborn as sns
plt.figure(figsize=(6,4))
sns.histplot(data=elbow)
plt.xlabel('Number of clusters k')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.plot(elbow, inertia, 'bo-')
plt.xlabel('Number of clusters k')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

```python
# let's choose k=3 based on elbow and fit

model=KMeans(n_clusters=3,random_state=1)
```

```python
df['cluster']=model.fit_predict(df)
```

```python
df
```

```python
plt.scatter(df['AnnualIncome'], df['SpendingScore'], c=df['cluster'], cmap='rainbow', s=100)
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title('Customer Segments')
plt.show()
```

```python
# Question 2: Use silhouette score to choose k (product features)

import pandas as pd
df = pd.DataFrame({
    'Price': [10,12,20,25,30,8,9,27,35,40,42,38,31,29,19],
    'UnitsSold': [200,180,150,120,100,220,210,130,90,80,75,85,95,105,160]
})
df
```

## 08-07-2025

```python
# compute silhouette score for k=2....5
# choose k with highest silhouette
# visualize

from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
scores = []
for k in range(2, 6):
    model = KMeans(n_clusters=k, random_state=0)
    labels = model.fit_predict(df)
    score = silhouette_score(df, labels)
    scores.append(score)
    print(f"k={k}, silhouette={score:.2f}")
```

```python
model=KMeans(n_clusters=2,random_state=0)
df['cluster']=model.fit_predict(df)
```

```python
df
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['Price'],df['UnitsSold'],c=df['cluster'],cmap='viridis',s=100)
plt.xlabel('Price')
plt.ylabel('Units Sold')
plt.title('Product Clusters')
plt.show()
```

```python
# Q6: Compare K=1,3,5 for same data (classification)
import pandas as pd
df = pd.DataFrame({
    'Feature1': [2,3,4,5,6,7],
    'Feature2': [8,7,6,5,4,3],
    'Class': [0,0,0,1,1,1]
})
df
```

```python
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
```

```python
x=df[['Feature1','Feature2']]
y=df['Class']
```

```python
plt.figure(figsize=(15,4))
for i, k in enumerate([1,3,5]):
    model=KNeighborsClassifier(n_neighbors=k)
    model.fit(x,y)
    pred=model.predict(x)

    plt.subplot(1,3,i+1)
    plt.scatter(df['Feature1'],df['Feature2'],c=pred,cmap='coolwarm',s=100,edgecolors='black')
    plt.title(f'K={k}')
    plt.xlabel('Feature1')
    plt.ylabel('Feature2')
plt.tight_layout()
plt.suptitle('KNN classification with different k values',y=1.05,fontsize=14)
plt.show()
```
