```python
salary_mapping={'low':1,'medium':2,'high':3}
salary_mapping
```

```python
df['salary_numeric']=df['salary'].map(salary_mapping)
```

```python
average=df.groupby('department')['salary_numeric'].mean()
```

```python
highest_salary=df.loc[df.groupby('department')['salary_numeric'].idxmax()]
```
